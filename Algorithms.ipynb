Linear Search

```python
def linear_search(arr,k):
    for i in arr:
        if i==k:
            return True
    return False
```

```python
arr = list(map(int,input("Enter the array elements : ").split()))
k = int(input("Enter the search element : "))
print(linear_search(arr,k))
```

Binary Search

```python
def binary_search(arr,k):
    mid = len(arr)//2
    if arr[mid] == k:
        return True
    if arr[mid] != k and len(arr)==1:
        return False
    if arr[mid] > k:
        return binary_search(arr[:mid],k)
    else:
        return binary_search(arr[mid+1:],k)
```

```python
arr = list(map(int,input("Enter the array elements : ").split()))
k = int(input("Enter the search element : "))
print(binary_search(arr,k))
```

Bubble Sort

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
    return arr
```

```python
arr = list(map(int,input("Enter the array elements : ").split()))
print(bubble_sort(arr))
```

Selection Sort

```python
def selection_sort(arr):
    for i in range(len(arr)):
        m,mi = arr[i],i
        for j in range(i+1,len(arr)):
            if arr[j] < m:
                m = arr[j]
                mi = j
        arr[mi] = arr[i]
        arr[i] = m
    return arr
```

```python
arr = list(map(int,input("Enter the array elements : ").split()))
print(selection_sort(arr))
```

Insertion Sort

```python
def insertion_sort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and arr[j]>key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
```

```python
arr = list(map(int,input("Enter the array elements : ").split()))
print(insertion_sort(arr))
```

Merge Sort

```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid = len(arr)//2
    left = arr[:mid]
    right = arr[mid:]
    
    left = merge_sort(left)
    right = merge_sort(right)
    
    return merge(left,right)

def merge(arr1,arr2):
    len_arr1, len_arr2 = len(arr1),len(arr2)
    i1,i2 = 0,0
    na = []
    while i1<len_arr1 and i2<len_arr2:
        if arr1[i1]<arr2[i2]:
            na.append(arr1[i1])
            i1+=1
        else:
            na.append(arr2[i2])
            i2+=1
    while i1<len_arr1:
        na.append(arr1[i1])
        i1+=1
    while i2<len_arr2:
        na.append(arr2[i2])
        i2+=1
    return na
```

```python
arr = list(map(int,input("Enter the array elements : ").split()))
print(merge_sort(arr))
```

Quick Sort

```python
def quick_sort(arr):
    if len(arr)==0:
        return []
    pivot = [arr[0]]
    sep = partition(arr)
    left, right = sep[0],sep[1]
    left = quick_sort(left)
    right = quick_sort(right)
    return left+pivot+right

def partition(arr):
    s = arr[0]
    l,r = [],[]
    for i in range(1,len(arr)):
        if arr[i]<=s:
            l.append(arr[i])
        else:
            r.append(arr[i])
    return [l,r]
```

```python
quick_sort([99,98,97,1,2,3])
```

Counting Sort

```python
def count_sort(arr):
    m = max(arr)+1
    new_arr = [0]*m
    for i in arr:
        new_arr[i] += 1
    i,j = 0,0
    while i < m:
        if new_arr[i] == 0:
            i += 1
            continue
        arr[j] = i
        j += 1
        new_arr[i] -= 1
    return arr
```

```python
count_sort([99,98,97,1,2,3])
```

Radix Sort

```python
def store(arr,curr,fin):
    if fin < curr:
        return arr
    new = [[]]*10
    for i in arr:
        t = i
        t = t%(10**curr)
        t = t//(10**(curr-1))
        new[t] = new[t]+[i]
    k = 0
    for i in new:
        for j in i:
            arr[k] = j
            k += 1
    return store(arr,curr+1,fin)

def radix_sort(arr):
    m,c = max(arr),0
    while m>0:
        c+=1
        m//=10
    return store(arr,1,c)
```

```python
radix_sort([121, 432, 564, 23, 1, 45, 788])
#radix_sort([99,98,97])
```

Shell Sort

```python
def shell_sort(arr):
    n = len(arr)
    interval = n//2
    while interval > 0:
        for i in range(interval,n):
            j = i
            temp = arr[i]
            while j>=interval and arr[j-interval]>temp:
                arr[j] = arr[j-interval]
                j-=interval
            arr[j] = temp
        interval//=2
    return arr
```

```python
shell_sort([121, 432, 564, 23, 1, 45, 788])
```
