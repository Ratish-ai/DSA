Stack

```python
class stack:
    
    def __init__(self):
        self.stack = []
    
    def push(self,n):
        self.stack.append(n)
    
    def pop(self):
        return self.stack.pop()
```

```python
n = int(input("Enter the length of the stack : "))
s = stack()
print("Enter the elements")
for i in range(n):
    s.push(int(input()))
for i in s.stack:
    print(i,end = " ")
print()
print(s.pop())
for i in s.stack:
    print(i,end = " ")
```

Queue

```python
class queue:
    
    def __init__(self):
        self.queue = []
    
    def enqueue(self,n):
        self.queue.append(n)
    
    def dequeue(self):
        return self.queue.pop(0)
```

```python
n = int(input("Enter the length of the Queue : "))
q = queue()
for i in range(n):
    q.enqueue(int(input()))
for i in q.queue:
    print(i,end=" ")
print()
print(q.dequeue())
for i in q.queue:
    print(i,end=" ")
```

Singly Linked List

```python
class node:
    
    def __init__(self,n):
        self.data = n
        self.next = None
```

```python
class singly:
    
    def __init__(self):
        self.head = None
    
    def add(self,n):
        new = node(n)
        dummy = self.head
        while dummy.next != None:
            dummy = dummy.next
        dummy.next = new
    
    def insert(self,pos,val):
        new = node(val)
        dummy = self.head
        if pos == 1:
            new.next = dummy
            self.head = new
        else:
            curr = 1
            while curr != pos-1:
                dummy = dummy.next
                curr += 1
            new.next = dummy.next
            dummy.next = new
        
    def delete(self,n):
        dummy = self.head
        if dummy.data == n:
            self.head = dummy.next
        else:
            while dummy.next.data != n:
                dummy = dummy.next
            dummy.next = dummy.next.next
    
    def show(self):
        sl = self.head
        while sl.next != None:
            print(sl.data,end=" -> ")
            sl = sl.next
        else:
            print(sl.data,end=" -> ")
        print(None)
```

```python
n = int(input("Enter Total no of elements : "))
print("Enter elements")
sl = singly()
for i in range(n):
    if i==0:
        sl.head = node(int(input()))
    else:
        sl.add(int(input()))
sl.show()
val = int(input("Enter the element to be inserted : "))
pos = int(input("Enter the position of the element : "))
sl.insert(pos,val)
sl.show()
val = int(input("Enter the element to be deleted : "))
sl.delete(val)
sl.show()
```

Doubly Linked List

```python
class node:
    
    def __init__(self,n):
        self.prev = None
        self.data = n
        self.next = None
```

```python
class doubly:
    
    def __init__(self):
        self.head = None
    
    def add(self,n):
        new = node(n)
        dummy = self.head
        while dummy.next != None:
            dummy = dummy.next
        dummy.next = new
        new.prev = dummy
    
    def printall(self):
        dummy = self.head
        while dummy.next!=None:
            print(dummy.data,end=' -> ')
            dummy = dummy.next
        print(dummy.data)
        while dummy.prev!=None:
            print(dummy.data,end = ' -> ')
            dummy = dummy.prev
        print(dummy.data)
    
    def insert(self,pos,val):
        dummy = self.head
        new = node(val)
        if pos==1:
            self.head.prev = new
            new.next = self.head
            self.head = new
            return
        curr = 1
        while curr<pos-1:
            dummy = dummy.next
            curr+=1
        if dummy.next==None:
            new.prev = dummy
            dummy.next = new
            return
        new.next = dummy.next
        new.prev = dummy
        dummy.next.prev = new
        dummy.next = new
    
    def delete(self,n):
        dummy = self.head
        if dummy.data==n:
            self.head = dummy.next
            self.head.prev = None
            return
        while dummy.data != n:
            dummy = dummy.next
        if dummy.data != n:
            return
        if dummy.next is None:
            dummy.prev.next = None
            return
        dummy.prev.next = dummy.next
        dummy.next.prev = dummy.prev
```

```python
n = int(input("Enter the no.of elements : "))
print("Enter the elements")
d = doubly()
for i in range(n):
    if i==0:
        d.head = node(int(input()))
    else:
        d.add(int(input()))
d.printall()
val = int(input("Enter a value to be inserted : "))
pos = int(input("Enter the position to be inserted : "))
d.insert(pos,val)
d.printall()
de = int(input("Enter the element to be deleted : "))
d.delete(de)
d.printall()
```

```python
class node:
    def __init__(self,n):
        self.left = None
        self.data = n
        self.right = None
```

```python
n = int(input("Enter the no.of elements : "))
bst = bst()
print("Enter the elements")
for i in range(n):
    if i==0:
        bst.head = node(int(input()))
    else:
        bst.add(int(input()))
bst.inorder(bst.head)
print("Minimum :",bst.minimum())
print("Maximum :",bst.maximum())
de = int(input("Enter the element to be deleted : "))
bst.delete(de)
de = int(input("Enter the element to be deleted : "))
bst.delete(de)
de = int(input("Enter the element to be deleted : "))
bst.delete(de)
```
