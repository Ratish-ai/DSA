Stack

```python
class stack:
    
    def __init__(self):
        self.stack = []
    
    def push(self,n):
        self.stack.append(n)
    
    def pop(self):
        return self.stack.pop()
```

```python
n = int(input("Enter the length of the stack : "))
s = stack()
print("Enter the elements")
for i in range(n):
    s.push(int(input()))
for i in s.stack:
    print(i,end = " ")
print()
print(s.pop())
for i in s.stack:
    print(i,end = " ")
```

Queue

```python
class queue:
    
    def __init__(self):
        self.queue = []
    
    def enqueue(self,n):
        self.queue.append(n)
    
    def dequeue(self):
        return self.queue.pop(0)
```

```python
n = int(input("Enter the length of the Queue : "))
q = queue()
for i in range(n):
    q.enqueue(int(input()))
for i in q.queue:
    print(i,end=" ")
print()
print(q.dequeue())
for i in q.queue:
    print(i,end=" ")
```

Singly Linked List

```python
class node:
    
    def __init__(self,n):
        self.data = n
        self.next = None
```

```python
class singly:
    
    def __init__(self):
        self.head = None
    
    def add(self,n):
        new = node(n)
        dummy = self.head
        while dummy.next != None:
            dummy = dummy.next
        dummy.next = new
    
    def insert(self,pos,val):
        new = node(val)
        dummy = self.head
        if pos == 1:
            new.next = dummy
            self.head = new
        else:
            curr = 1
            while curr != pos-1:
                dummy = dummy.next
                curr += 1
            new.next = dummy.next
            dummy.next = new
        
    def delete(self,n):
        dummy = self.head
        if dummy.data == n:
            self.head = dummy.next
        else:
            while dummy.next.data != n:
                dummy = dummy.next
            dummy.next = dummy.next.next
    
    def show(self):
        sl = self.head
        while sl.next != None:
            print(sl.data,end=" -> ")
            sl = sl.next
        else:
            print(sl.data,end=" -> ")
        print(None)
```

```python
n = int(input("Enter Total no of elements : "))
print("Enter elements")
sl = singly()
for i in range(n):
    if i==0:
        sl.head = node(int(input()))
    else:
        sl.add(int(input()))
sl.show()
val = int(input("Enter the element to be inserted : "))
pos = int(input("Enter the position of the element : "))
sl.insert(pos,val)
sl.show()
val = int(input("Enter the element to be deleted : "))
sl.delete(val)
sl.show()
```

Doubly Linked List

```python
class node:
    
    def __init__(self,n):
        self.prev = None
        self.data = n
        self.next = None
```

```python
class doubly:
    
    def __init__(self):
        self.head = None
    
    def add(self,n):
        new = node(n)
        dummy = self.head
        while dummy.next != None:
            dummy = dummy.next
        dummy.next = new
        new.prev = dummy
    
    def printall(self):
        dummy = self.head
        while dummy.next!=None:
            print(dummy.data,end=' -> ')
            dummy = dummy.next
        print(dummy.data)
        while dummy.prev!=None:
            print(dummy.data,end = ' -> ')
            dummy = dummy.prev
        print(dummy.data)
    
    def insert(self,pos,val):
        dummy = self.head
        new = node(val)
        if pos==1:
            self.head.prev = new
            new.next = self.head
            self.head = new
            return
        curr = 1
        while curr<pos-1:
            dummy = dummy.next
            curr+=1
        if dummy.next==None:
            new.prev = dummy
            dummy.next = new
            return
        new.next = dummy.next
        new.prev = dummy
        dummy.next.prev = new
        dummy.next = new
    
    def delete(self,n):
        dummy = self.head
        if dummy.data==n:
            self.head = dummy.next
            self.head.prev = None
            return
        while dummy.data != n:
            dummy = dummy.next
        if dummy.data != n:
            return
        if dummy.next is None:
            dummy.prev.next = None
            return
        dummy.prev.next = dummy.next
        dummy.next.prev = dummy.prev
```

```python
n = int(input("Enter the no.of elements : "))
print("Enter the elements")
d = doubly()
for i in range(n):
    if i==0:
        d.head = node(int(input()))
    else:
        d.add(int(input()))
d.printall()
val = int(input("Enter a value to be inserted : "))
pos = int(input("Enter the position to be inserted : "))
d.insert(pos,val)
d.printall()
de = int(input("Enter the element to be deleted : "))
d.delete(de)
d.printall()
```

Binary Search Tree

```python
class node:
    def __init__(self,n):
        self.left = None
        self.data = n
        self.right = None
```

```python
class bst:
    def __init__(self):
        self.head = None
    
    def add(self,n,dummy = None):
        if dummy is None:
            dummy = self.head
        if n>dummy.data and dummy.right is None:
            new = node(n)
            dummy.right = new
            return
        elif n<=dummy.data and dummy.left is None:
            new = node(n)
            dummy.left = new
            return
        if n>dummy.data:
            return self.add(n,dummy.right)
        else:
            return self.add(n,dummy.left)
    
    def minimum(self,node = None):
        if node is None:
            node = self.head
        if node.left is None:
            return node.data
        return self.minimum(node.left)
    
    def maximum(self,node = None):
        if node is None:
            node = self.head
        if node.right is None:
            return node.data
        return self.maximum(node.right)
    
    def delete(self,n,node = None):
        if node is None:
            node = self.head
        if node.data==n:
            if node.left is None and node.right is None:
                # leaf
                print("leaf")
                return None
            elif node.left is None:
                # right child
                print("only right")
                temp = node.right
                node = None
                return temp
            elif node.right is None:
                # left child
                print("only left")
                temp = node.left
                node = None
                return temp
            else:
                # 2 child
                print("2 child")
                new = self.minimum(node.right)
                node.data = new
                node.right = self.delete(new,node.right)
                return node
        elif n>node.data:
            node.right = self.delete(n,node.right)
        else:
            node.left = self.delete(n,node.left)
        return node
    
    def inorder(self,dummy):
        if dummy:
            self.inorder(dummy.left)
            print(dummy.data,end = ' ')
            self.inorder(dummy.right)
```

```python
n = int(input("Enter the no.of elements : "))
bst = bst()
print("Enter the elements")
for i in range(n):
    if i==0:
        bst.head = node(int(input()))
    else:
        bst.add(int(input()))
bst.inorder(bst.head)
print()
print("Minimum :",bst.minimum())
print("Maximum :",bst.maximum())
de = int(input("Enter the element to be deleted : "))
bst.delete(de)
bst.inorder(bst.head)
print()
```

Graph

```python
class graph:
    def __init__(self,vertex):
        self.adj_mat = [[0 for j in range(vertex)] for i in range(vertex)]
        self.size = vertex
    
    def add_edge(self,v1,v2):
        if v1==v2:
            raise ValueError()
        elif self.adj_mat[v1][v2]==1:
            raise ValueError()
        else:
            self.adj_mat[v1][v2] = 1
            self.adj_mat[v2][v1] = 1
    
    def del_edge(self,v1,v2):
        if v1==v2:
            raise ValueError()
        elif self.adj_mat[v1][v2]==0:
            raise ValueError()
        else:
            self.adj_mat[v1][v2] = 0
            self.adj_mat[v2][v1] = 0
    
    def show(self):
        for i in self.adj_mat:
            print(*i)
```

```python
n = int(input('Enter the no of nodes in the graph : '))
e = int(input('Enter the no of edges : '))
g = graph(n)
g.show()
c = 0
while c<e:
    try:
        v1,v2 = map(int,input("Enter the edges to connect : ").split())
        g.add_edge(v1,v2)
    except:
        print('Both vertex are same or Edge already exists')
        continue
    c+=1
g.show()
try:
    v1,v2 = map(int,input("Enter the edges to delete : ").split())
    g.del_edge(v1,v2)
except:
    print("Both vertex are same or Edge doesn't exists")
else:
    g.show()
```

```python
class node:
    def __init__(self,vertex):
        self.value = vertex
        self.next = None

class graph:
    def __init__(self,n):
        self.size = n
        self.graph = [None]*n
    
    def add_edge(self,v1,v2):
        new_node = node(v2)
        new_node.next = self.graph[v1]
        self.graph[v1] = new_node
        
        new_node = node(v1)
        new_node.next = self.graph[v2]
        self.graph[v2] = new_node
    
    def del_edge(self,v1,v2):
        self.graph[v1] = self.del_node(self.graph[v1],v2)
        self.graph[v2] = self.del_node(self.graph[v2],v1)
    
    def del_node(self,node_arr,vert):
        if vert==node_arr.value:
            return node_arr.next
        temp = node_arr.next
        while temp.next.value != vert:
            if temp is None:
                return node_arr
            temp = temp.next
        if temp.next.value==vert:
            temp.next = temp.next.next
        return node_arr
    
    def show(self):
        for i in range(self.size):
            print(f"Vertex {i}: ", end="")
            temp = self.graph[i]
            while temp:
                print(f" -> {temp.value}", end="")
                temp = temp.next
            print()
```

```python
n = int(input('Enter the no of nodes in the graph : '))
e = int(input('Enter the no of edges : '))
g = graph(n)
g.show()
for i in range(e):
    v1,v2 = map(int,input("Enter the edges to connect : ").split())
    g.add_edge(v1,v2)
g.show()
v1,v2 = map(int,input("Enter the edges to delete : ").split())
g.del_edge(v1,v2)
g.show()
```
